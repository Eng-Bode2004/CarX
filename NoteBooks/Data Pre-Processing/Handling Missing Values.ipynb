```python
import pandas as pd
Path = r'/Datasets/Raw/Cars.csv'
Cars = pd.read_csv(Path)
```

```python
Cars.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(Cars.isnull())
```

```python
Cars['mileage(km/ltr/kg)'].fillna(Cars['mileage(km/ltr/kg)'].mean(), inplace=True)
Cars.isnull().sum()
Cars.to_csv(Path)
```

```python
Cars['seats'] = Cars['seats'].fillna(5.0)  # Replace missing values with 5
Cars.to_csv(Path, index=False)  # Save the updated dataset

```

```python
Cars['max_power'] = Cars['max_power'].fillna(74)  # Replace missing values with 5
Cars.to_csv(Path, index=False)  # Save the updated dataset
```
